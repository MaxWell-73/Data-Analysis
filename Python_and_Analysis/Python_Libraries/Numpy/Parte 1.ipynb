```python
df = pd.read_csv('/kaggle/input/fifa23-official-datasetclean-data/CLEAN_FIFA23_official_data.csv')
```

```python
df.tail()
```

```python
world_champions = ['Argentina','Italy','England','Uruguay']
df_dict = {}

for champ in world_champions:
    df_dict[champ] = df[df['Nationality'] == champ]
```

Primeros 7 jugadores que aparecen en la lista de cada pais

```python
jugadores = {}
for country in df_dict:
    jugadores[country] = df_dict[country]['Name'][:7]
```

```python
jugadores
```

```python
est_overall = {}
for country in df_dict:
    est_overall[country] = [ {'minimo': np.min(df_dict[country]['Overall'])},
                            {'maximo': np.max(df_dict[country]['Overall'])},
                            {'promedio': np.mean(df_dict[country]['Overall'])},
                            {'Desviacion Estandar': np.std(df_dict[country]['Overall'])}]  
    
```
