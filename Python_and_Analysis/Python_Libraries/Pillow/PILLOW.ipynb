# Tutorial de PILLOW

La clase más importante en esta librería es la clase `Image`, la cual se usa para cargar imágenes, procesarlas o incluso crearlas desde cero.
[PILLOW](https://pillow.readthedocs.io/en/stable/handbook/concepts.html)

```python
import PIL
from PIL import Image, ImageFilter, ImageEnhance
import urllib.request
import matplotlib.pyplot as plt
```

## Cargando una imagen desde una URL

```python
urllib.request.urlretrieve('https://github.com/MaxWell-73/Data-Analysis/blob/main/Python_and_Analysis/Python_Libraries/Pillow/gato.png?raw=true', 'file_name.png')
img = Image.open('file_name.png')
img.show()
```

```python
img.format  # formato de la imágen (.png .jpg .jpeg .gif .webp ...)
```

```python
img.size  # ancho y alto de la imágen en pixeles
```

```python
img.mode  # base de los colores de la imágen (1, L, P, RBG, RGBA, CMYK, YCbCr)
```

## Cambiando el formato del color de la imágen

```python
modes = ['1', 'L', 'RGB', 'RGBA', 'CMYK', 'YCbCr']
fig, ax = plt.subplots(1, 6, figsize=(12, 20))
for i in range(6):
  img_new = img.convert(modes[i])  # pasamos el color de la imágen a formato RGB
  ax[i].imshow(img_new, cmap = 'gray' if modes[i] == 'L' else None)
  ax[i].axis('off')
  ax[i].set_title(f'Mode {modes[i]}')
plt.show()
```

```python
img = img.convert('RGB')  # lo dejamos en formato RGB
img.show()
```

## Separar en canales R G B la imagen



```python
r, g , b = img.split()

fig, ax = plt.subplots(1,3,figsize=(12,20))

ax[0].imshow(r,cmap = 'Reds')
ax[1].imshow(g,cmap = 'Greens')
ax[2].imshow(b,cmap = 'Blues')

plt.show()
```

```python
im = Image.merge("RGB", (r, g, b)) # se pueden unir imagenes o sus canales para formar una nueva imagen.
im.show()
```

```python
im = Image.merge("RGB", (b, g, r)) # Es sensible al orden de la tupla (r, g, b)
im.show()
```

## Modifificar imagenes (cortar, rotar, superponer, cambiar de tamaño)

```python
img = img.rotate(30)
img.show()
```

```python
box = (100, 100, 400, 400) # Se recorta con coordenadas (x_init, y_init, x_end, y_end)
region = img.crop(box)

region.show()
```

```python
transposed_image = region.transpose(Image.TRANSPOSE)
# transposed_image = transposed_image.rotate(30) # se puede rotar antes de sobreponer las imagenes
img.paste(transposed_image, box)
```

```python
img = Image.open('file_name.png')
img.resize((500, 500))
```

```python
img = Image.open('file_name.png')
img = img.convert('RGB')
out = img.filter(ImageFilter.DETAIL) # los filtros son distintos https://pillow.readthedocs.io/en/stable/handbook/concepts.html

fig, ax = plt.subplots(1,2, figsize=(10,12))

ax[0].imshow(img)
ax[1].imshow(out)
```

```python

enh = ImageEnhance.Contrast(img)
enh.enhance(1.3).show("30% more contrast")
```
