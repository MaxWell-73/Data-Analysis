# **Create SQLite databases from Python**

```python
import sqlite3

connection = sqlite3.connect("colombian_league.db")
cursor = connection.cursor()

cursor.execute("create table champions (year integer, team text, city text)")
```

```python
champions_list = [
    (20231,"Millonarios",None),
    (20222,"Deportivo Pereira",None),
    (20221,"Atletico Nacional",None),
    (20212,"Deportivo Cali",None),
    (20212,"Deportes Tolima",None)
]

cursor.executemany("insert into champions values (?,?,?)", champions_list)

# Print database row
for row in cursor.execute("select * from champions"):
  print(row)

print("****************************")
#Print specific rows
cursor.execute("select * from champions where team=:c",{"c":"Millonarios"})
league_search = cursor.fetchall()
print(league_search)

cities_list = [
    ("Millonarios", "Bogotá"),
    ("Atletico Nacional", "Medellín"),
    ("Deportivo Pereira", "Pereira"),
    ("Deportivo Cali", "Cali")
]

cursor.execute("create table cities (team text, city text)")
cursor.executemany("insert into cities values (?,?)", cities_list)
cursor.execute("select * from cities where team=:c",{"c":"Millonarios"})
cities_search = cursor.fetchall()
print(cities_search)

print("*************************")


#for i in league_search:
for i in league_search:
  adjusted = [cities_search[0][1] if value == cities_search[0][0] else value for value in i]
  print(adjusted)
connection.close()
```
