```python
rightFoot = (FIFA23['Overall']>=85) & (FIFA23['Preferred Foot']=='Right')
rf = FIFA23[rightFoot].shape[0] #shape devuelve el numero de filas y el numero de columnas

leftFoot = (FIFA23['Overall']>=85) & (FIFA23['Preferred Foot']=='Left')
lf = FIFA23[leftFoot].shape[0] #shape devuelve el numero de filas y el numero de columnas

total85 = FIFA23[FIFA23['Overall']>=85].shape[0]
print('Pierna derecha:',rf,'\n','Pierna izquierda:',lf,'\n','Total:',total85)
```

```python
categorias = ['Pierna Derecha', 'Pierna Izquierda']
valores = [71, 23]

# Crear el gráfico de torta
plt.pie(valores, labels=categorias, autopct='%1.1f%%')

# Ajustar el aspecto del gráfico
plt.axis('equal')

# Mostrar el gráfico
plt.show()
```
